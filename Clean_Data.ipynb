# Data Merging

```python
import pandas as pd
import numpy as np
```

```python
# reading ACS Socio-Demographics Curation_2023
xls = pd.ExcelFile('/content/drive/MyDrive/KPMG #1 (NYC) - Realizing Gender Equality in the Workplace: Low Wage Geographies (Team B)/Data/KPMG_data_files/ACS Socio-Demographics Curation_2023.xlsx')

df_curated = pd.read_excel(xls, 'ACS Socio-Demographics Curation')
df_key = pd.read_excel(xls, 'Data Dictionary')
```

```python
df_curated.columns = df_curated.columns.str.upper()

#map table columns from dfKey
column_mapping = dict(zip(df_key["Key"], df_key['Label']))
df_curated.rename(columns=column_mapping, inplace=True)

df_curated.to_csv("curated.csv", index=False)
curated = pd.read_csv("curated.csv")
curated.rename({'ZCTA': 'ZCTA5'}, axis=1, inplace=True)
curated.head()
```

```python
df_gender = pd.read_excel("/content/drive/MyDrive/KPMG #1 (NYC) - Realizing Gender Equality in the Workplace: Low Wage Geographies (Team B)/Data/KPMG_data_files/ACS 2021 Sociodemographics by Sex.xlsx")
df_gender.shape
```

```python
df_gender.columns = df_gender.columns.str.upper()
```

```python
#merging on ZCTA5
combined_df = pd.merge(curated, df_gender, on="ZCTA5")
```

```python
combined_df.to_csv("merged.csv", index=False)
merged = pd.read_csv("merged.csv")
merged.head()
```

# Data Filtering

```python
# Selecting only NY data
merged = merged[merged['STATE'] == 'NY']
merged.shape
```

```python
# NYC boroughs in county FIPS code
borough_val = [36005, 36047, 36061, 36081, 36085]
borough_name = ["Bronx", "Brooklyn", "Manhattan", "Queens", "Staten_Island"]

# Get boroughs
merged = merged[merged['COUNTY_FIPS'].isin(borough_val)]

# Change values in COUNTY FIPS to its string value
merged['BOROUGH'] = merged['COUNTY_FIPS'].replace(borough_val, borough_name)

#Make borough the index
merged = merged.set_index('BOROUGH')

merged.shape
```

```python
# Count num of data for each borough
print("Manhattan", merged['COUNTY_FIPS'].value_counts()[36061])
print("Brooklyn", merged['COUNTY_FIPS'].value_counts()[36047])
print("Queens", merged['COUNTY_FIPS'].value_counts()[36081])
print("Bronx", merged['COUNTY_FIPS'].value_counts()[36005])
print("Staten Island", merged['COUNTY_FIPS'].value_counts()[36085])
```

```python
# Count the number of zeros in each row
zeros_count = merged.eq(0).sum(axis=1)

# Set a threshold for the majority of zeros
threshold = merged.shape[1] // 2

# Filter rows where the number of zeros is less than the threshold
merged = merged[zeros_count < threshold]

merged.shape
```

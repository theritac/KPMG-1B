```python
def get_borough(row):
    lat = row['LOCATION_LAT']
    lon = row['LOCATION_LON']

    if (40.7002 <= lat <= 40.8785) and (-74.0479 <= lon <= -73.9067):
        return 'Manhattan'
    elif (40.5707 <= lat <= 40.7394) and (-74.0418 <= lon <= -73.8499):
        return 'Brooklyn'
    elif (40.4774 <= lat <= 40.8007) and (-73.9626 <= lon <= -73.7004):
        return 'Queens'
    elif (40.7855 <= lat <= 40.9176) and (-73.9332 <= lon <= -73.7654):
        return 'Bronx'
    elif (40.4774 <= lat <= 40.6482) and (-74.2591 <= lon <= -74.0341):
        return 'Staten Island'
    else:
        return 'NOT NYC'
```

```python
named_df_table1['BOROUGH'] = named_df_table1.apply(get_borough, axis=1)
```

## Clean Up Columns & Data analysis

```python
named_df_table1['BOROUGH'] = named_df_table1.apply(get_borough, axis=1)

#drop non-NYC data
named_df_table1.drop(named_df_table1[named_df_table1['BOROUGH'] == 'NOT NYC'].index, inplace=True)

#drop columns where >=90% is NaN
named_df_table1 = named_df_table1.loc[:, named_df_table1.isnull().mean() <= .9]

#Make borough the index
named_df_table1 = named_df_table1.set_index('BOROUGH')
```

```python
poverty_status = named_df_table1.loc[:, named_df_table1.columns.str.startswith('POVERTY STATUS')]
poverty_status = poverty_status.groupby(['BOROUGH']).sum()
```

```python
pd.DataFrame(poverty_status)
```

```python
plt.bar(poverty_status.index, poverty_status['POVERTY STATUS IN THE PAST 12 MONTHS BY SCHOOL ENROLLMENT BY LEVEL OF SCHOOL FOR THE POPULATION 3 YEARS AND OVER%Population 3 years and over for whom poverty status is determined%Total%Income in the past 12 months below the poverty level']/poverty_status.sum(axis=1) * 100)
plt.xlabel('Boroughs')
plt.ylabel('Income in the past 12 months below the poverty level')
plt.tight_layout()
plt.show()
```

```python
named_df_table2['BOROUGH'] = named_df_table2.apply(get_borough, axis=1)

#drop non-NYC data
named_df_table2.drop(named_df_table2[named_df_table2['BOROUGH'] == 'NOT NYC'].index, inplace=True)

#drop columns where >=90% is NaN
named_df_table2 = named_df_table2.loc[:, named_df_table2.isnull().mean() <= .9]

#Make borough the index
named_df_table2 = named_df_table2.set_index('BOROUGH')
```

```python
employment_status = named_df_table2.loc[:, named_df_table2.columns.str.startswith('SEX BY AGE BY EMPLOYMENT STATUS FOR THE POPULATION 16 YEARS AND OVER')]
employment_status = employment_status.groupby(['BOROUGH']).sum()
```

```python
pd.DataFrame(employment_status)
```

```python
employment_status['Difference'] = abs(employment_status['SEX BY AGE BY EMPLOYMENT STATUS FOR THE POPULATION 16 YEARS AND OVER%Population 16 years and over%Total%Male']/employment_status.sum(axis=1) - employment_status['SEX BY AGE BY EMPLOYMENT STATUS FOR THE POPULATION 16 YEARS AND OVER%Population 16 years and over%Total%Female']/employment_status.sum(axis=1)) * 100
```

```python
employment_status['Difference']
```

```python
plt.bar(employment_status.index, employment_status['Difference'])
plt.ylabel('Difference of employement status by sex')
plt.tight_layout()
plt.show()
```
